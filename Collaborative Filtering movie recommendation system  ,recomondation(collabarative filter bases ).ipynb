```python
#transform data frame;=>
#df.pivot_table(index,column,values)   : transform table into  new formate table vector form
#___________________________________


user_movie_df=df.pivot_table(index='userId',columns='movieId',values='rating')
```

```python
model=NearestNeighbors(metric="cosine")
model.fit(user_movie_df.values)    #values;only show  df values in vector form
```

```python
#=>#recommand movie for user id 122

import pandas as pd
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric="cosine")
model.fit(user_movie_df.values)

df=pd.read_csv('ML_Dataset/movies_collaborative.csv')
id_title=df[['movieId','title']]
id_title=id_title.drop_duplicates()

user_movie_df=df.pivot_table(index='userId',columns='movieId',values='rating')
user_movie_df.fillna(0,inplace=True)

target_userId=122                                            #recommand movie for user id 122
target_userMId=user_movie_df.loc[target_userId]
scores,indexes=model.kneighbors([target_userMId],n_neighbors=2)
similar_userId=indexes[0][1]+1
similar_userMId=user_movie_df.loc[similar_userId]

target_set=set()
for mid,rating in zip(target_userMId.index,target_userMId):
    if rating>0:
        target_set.add(mid)

similar_set=set()
for mid,rating in zip(similar_userMId.index,similar_userMId):
    if rating>0:
        similar_set.add(mid)

recommed_mvs=similar_set-target_set

similar_dict={}
for mid,rating in zip(similar_userMId.index,similar_userMId):
    if mid in recommed_mvs and rating>4:
        similar_dict[mid]=rating
        
id_title[id_title.movieId.isin(similar_dict.keys())].head(3)
```
